<h1>Завантажити набір даних для регресії на <a href="https://www.kaggle.com/datasets?search=regression">сайті</a></h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
market_info = pd.read_csv('yahoo_stock.csv')




market_info.plot.scatter("Volume","Adj Close")
market_info

```

```python
import sklearn
from sklearn.model_selection import train_test_split

columns=market_info.columns

x = market_info[["High","Low","Open","Close","Adj Close"]]
y = market_info[["Volume"]] * (10 **(-9))
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_Hat = lr.predict(x_test)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error 
mean_squared_error(y_test, y_Hat),mean_absolute_error(y_test, y_Hat)
plt.plot(x_test,y_test, 'r.')
plt.plot(x_test,y_Hat, 'b.')

plt.show()
```
