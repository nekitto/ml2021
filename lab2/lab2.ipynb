<h1>Reading data</h1>

```python
import numpy as np
import matplotlib.pyplot as plt


market_info = pd.read_csv('yahoo_stock.csv')

market_info
```

<h1>Splitting,fitting,predicting</h1>

```python
import sklearn
from sklearn.model_selection import train_test_split


x = market_info[["High","Low","Open","Close"]]
y = market_info[["Volume"]] * (10 **(-9))
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.1)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_Hat = lr.predict(x_test)

fig, ax = plt.subplots()
ax.plot(y_test, 'r.', label = 'tested value')
ax.plot(y_Hat, 'b.', label = 'predicted value')
fig.set_figheight(10)
fig.set_figwidth(30)
ax.legend()
plt.show()
```
